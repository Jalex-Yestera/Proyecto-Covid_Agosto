```python

# Importamos bibliotecas

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
# Guardamos la url para poder actualizar.

url="https://covid.ourworldindata.org/data/owid-covid-data.csv"
```

```python
# Convertimos el csv en dataframe df y establecemos que "date" es tipo fecha y "location" el índice. Estudiamos las columnas.

df=pd.read_csv(url, parse_dates=["date"], index_col="location")
df.info()
```

```python
# Comprobamos si hay valores negativos en las columnas seleccionadas

neg=df.iloc[:,3:22]
neg[neg<0].sum()
```

```python
# Para buscar valores negativos en cada columna

neg[neg["new_cases"]<0]
```

```python
neg[neg["new_deaths"]<0]
```

# Conclusiones:

### Varios países tienen valores negativos para ajustar errores. Para nuestro estudio, observamos valores negativos en columnas relacionadas con nuevos casos de España y Francia y en columnas relacionadas con nuevas muertes en España.

# POPULATION

```python
# Selecionamos las columnas relativas a la población para ver los datos

pop=df.iloc[:,25:27]
pop=pop.T
pop=pop.loc[:,["France","India","Peru","Spain","United States"]].T

pop=pop[~pop.duplicated()]
popp=pop["population"]
popp
pop
```

```python
# Hacemos gráfico para ver qué ratio d población tiene cada uno de nuestros países sobre el conjunto.

colores_paises = ["dodgerblue","red","g","magenta","darkorange"]

def autopct_fun(abs_values):
    gen = iter(abs_values)
    return lambda pct: f"{pct:.1f}% ({next(gen)})"

counts = popp
popp.plot(kind="pie",autopct=autopct_fun(counts),colors=colores_paises,shadow=True,startangle=90,radius=2)
plt.savefig("../resources/plots/paises_juntos/population_pie.png")
plt.show()
plt.close()
```

```python
# Gráfico para ver que puesto ocupa cada país respecto a la variable "population"

popp.plot(kind="bar",color=colores_paises)
plt.title("POPULATION")
plt.savefig("../resources/plots/paises_juntos/population_bar.png")
plt.show()
plt.close()
```
